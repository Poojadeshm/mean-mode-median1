```python
# mean
numbers = [11, 96, 33, 54, 85]
total = 0

for num in numbers:
   total = total+ num
mean = total / len(numbers)
print("Mean:", mean)
```

```python
#median
numbers = [11, 96, 33, 54, 85]
numbers.sort() 
n = len(numbers)
if n % 2 == 0:
   median = (numbers[n//2 - 1] + numbers[n//2]) / 2
else:
   median = numbers[n//2]
print("Median:", median)
```

```python
# mode
numbers = [11, 96, 11, 54, 85,]
frequency = {}
# Count occurrences of each number
for num in numbers:
   if num in frequency:
       frequency[num] += 1
   else:
       frequency[num] = 1
max_count = max(frequency.values())
modes = [key for key, value in frequency.items() if value == max_count]
if len(modes) == len(numbers):
   print("No mode found")
else:
   print("Mode:", modes)
```
